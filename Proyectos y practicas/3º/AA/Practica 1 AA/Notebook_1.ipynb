#### Vamos a generar un nuevo df con las variables que tengan una correlacion menor a 0.9

Vamos a intentar eliminar las variables que esten muy correlacionadas entre si, para reducir la dimensionalidad del problema haciendo que el modelo sea mas interpretable y mas rapido. Asi como evitar problemas de multicolinealidad.

Se ha tanteado con el umbral para intentar balancear la precisión del modelo con la reducción de dimensionalidad, y se ha decidido que 0.95 es un buen umbral para este caso.

```python

# Identificar pares de variables con correlación >= 0.9 (excluyendo la diagonal)
high_corr_pairs = (correlation.where(np.triu(np.ones(correlation.shape), k=1).astype(bool))
                   .stack()
                   .where(lambda x: x >= 0.95)
                   .dropna())

energy_corr_with_vars = dict(correlation['energy'].drop('energy'))

# Determinar qué variables eliminar basándose en la correlación menor con 'energy'
vars_to_remove = []
for pair in high_corr_pairs.index:
    # Comparar la correlación de cada variable del par con 'energy'
    if energy_corr_with_vars[pair[0]] < energy_corr_with_vars[pair[1]]:
        vars_to_remove.append(pair[0])
    else:
        vars_to_remove.append(pair[1])

# Eliminar variables duplicadas en la lista de eliminación
vars_to_remove = list(set(vars_to_remove))

# Ahora, vamos a asegurarnos de que eliminamos correctamente las variables, incluso si hay correlaciones perfectas.
# Primero, crearemos una copia del DataFrame original para no modificar el existente mientras iteramos.
correlation_reduced = correlation.copy()

# Eliminaremos cada variable en vars_to_remove del DataFrame copiado.
for var in vars_to_remove:
    if var in correlation_reduced.columns:
        correlation_reduced.drop(var, axis=1, inplace=True)
        correlation_reduced.drop(var, axis=0, inplace=True)  # También eliminamos la fila para mantener la matriz cuadrada

wind_ava = wind_ava.drop(vars_to_remove, axis=1)

# Finalmente, mostraremos las variables que se han eliminado y la matriz de correlación reducida.
print(f"Variables eliminadas: {vars_to_remove}")
display(correlation_reduced.head())

# Mostrar la matriz de correlación reducida
plt.figure(figsize=(15, 15))
sns.heatmap(correlation_reduced, annot=True, cmap="GnBu", fmt=".2f")
plt.title("Matriz de correlación reducida sin variables redundantes")
plt.show()
```

Podemos ver que se ha eliminado 7 variablers y aun que hay variables altamente relacionadas entre si todavia (fsr y flsr, stl2 y stl3, etc), se ha reducido la dimensionalidad del problema sustancialmente y ya no hay variables con correlaciones mayores a 0.95.

#### Eliminamos las variables que no aportan información
Vamos a eliminar las variables que no tienen correlación con la variable objetivo, para intentar reducir la dimensionalidad del problema y hacer que el modelo sea mas interpretable y mas rapido.
De nuevo hemos hecho un balance entre la precisión del modelo y la reducción de dimensionalidad, y hemos decidido que 0.01 es un buen umbral para este caso.

```python
# Eliminamos las variables que tengan una correlación menor a 0.05 con la variable dependiente energy.
low_corr_vars = correlation_reduced['energy'][correlation_reduced['energy'] < 0.01].index
wind_ava = wind_ava.drop(low_corr_vars, axis=1)
print("eliminadas ", len(low_corr_vars), " variables con correlación menor a 0.01 con la variable energy")
# Volveremos a mostrar la matriz de correlación para verificar que las variables se han eliminado correctamente.
correlation = abs(wind_ava.corr())
plt.figure(figsize=(15, 15))
sns.heatmap(correlation, annot=True, cmap="GnBu", fmt=".2f")
plt.title("Matriz de correlación de wind_ava")
plt.show()
```

## Sistemas de evaluación

### Evaluación Outer e Inner, seleccion de las medidas del Error

Como hemos podido observar en el EDA, los datos contienen una gran cantidad de datos atipicos, que podemos suponer que hay ruido en los datos, es por ello no podemos utilizar unicamente MSE como medida pues es muy probable que sobreestime el error.
Ademas los valores de energy van desde 0 en la mayoria de casos a varios miles en otros, lo cual tambien es un problema con el MSE, por lo que podriamos plantearnos emplear MPSE, aun asi debido a los valores atipicos esta medida tampoco nos parece la ideal. 
Lo mismo ocurre aun que en menor medida con RMSE Y RMSPE.
Una medida que poriamos plantearnos usar es MAPE o MAE para solucionar este problema con los valores atipicos, pero nos introduce de nuevo conflicto con el gran rango de valores que toma cape.13.

Teniendo esto en cuenta consideramos que las mejores medidas del error para este caso son:
- R2
- **RMSE**
- MAE

#### Creación de un modelo dummy

Basandonos en las medidas seleccionadas anteriormente vamos a crear un regresor dummy y evaluarlo.

```python

from sklearn.dummy import DummyRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Primero, dividiremos los datos en entrenamiento y test sin varaibles correlacionadas
X = wind_ava.drop(columns='energy')
y = wind_ava['energy']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

# Entrenar un modelo de regresión "dummy" que prediga la media
# Este modelo es útil como referencia para comparar con otros modelos en el futuro
dummy = DummyRegressor(strategy="mean")
dummy.fit(X_train, y_train)

def rmse(y_true, y_pred):
    return np.sqrt(metrics.mean_squared_error(y_true, y_pred))

dummy_rmse = rmse(y_test, dummy.predict(X_test))
dummy_r2 = metrics.r2_score(y_test, dummy.predict(X_test))
dummy_mae = metrics.mean_absolute_error(y_test, dummy.predict(X_test))

print(f"RMSE del modelo dummy: {dummy_rmse:.2f}")
print(f"R^2 del modelo dummy: {dummy_r2:.2f}")
print(f"MAE del modelo dummy: {dummy_mae:.2f}")
```

Como se puede observar, haciendo uso de dos modelos de regresión "dummy" se tiene a un error cuadrático medio elevado.
Este modelo lo usaremos para comparar con nuestro modelo real

## Metodo de Escalado con KNN

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, StandardScaler

#normalizacion por escala min-max
min_max = make_column_transformer(
    (MinMaxScaler(), X_train.columns)
)

#normalizacion por escala estandar
standard = make_column_transformer(
    (StandardScaler(), X_train.columns)
)
```

### Escalado min-max

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
import time

pipe_knn = Pipeline([
        ('preproceso', min_max), #entrada del pipeline
        ('regresor', KNeighborsRegressor()) #salida del pipeline
])

np.random.seed(SEED)

t1 = time.time()
pipe_knn.fit(X_train, y_train)
t2 = time.time()

y_test_pred = pipe_knn.predict(X_test)

rmse_knn = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
r2_knn = metrics.r2_score(y_test, y_test_pred)
mae_knn = metrics.mean_absolute_error(y_test, y_test_pred)
print(f"RMSE: {rmse_knn}")
print(f"R2: {r2_knn}")
print(f"MAE: {mae_knn}")
print("Tiempo de entrenamiento: ", t2 - t1)
```

### Escalado estandar

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
import time

pipe_knn = Pipeline([
        ('preproceso', standard), #entrada del pipeline
        ('regresor', KNeighborsRegressor()) #salida del pipeline
])

np.random.seed(SEED)

t1 = time.time()
pipe_knn.fit(X_train, y_train)
t2 = time.time()

y_test_pred = pipe_knn.predict(X_test)

rmse_knn = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))
r2_knn = metrics.r2_score(y_test, y_test_pred)
mae_knn = metrics.mean_absolute_error(y_test, y_test_pred)
print(f"RMSE: {rmse_knn}")
print(f"R2: {r2_knn}")
print(f"MAE: {mae_knn}")
print("Tiempo de entrenamiento: ", t2 - t1)
```

El escalado estandar es el que mejor se ajusta a los datos, por lo que lo usaremos en el resto del notebook

### Conclusiones
- **¿Cuál es el mejor método?**
El mejor método es Random Forest, que ofrece un equilibrio óptimo entre precisión y tiempo de entrenamiento. Aunque modelos más simples como la regresión lineal y KNN son más rápidos, Random Forest supera a estos en términos de precisión, especialmente después de ajustar sus hiperparámetros.


- **¿Cuál es el mejor método de optimización de hiperparámetros?**
Entre los métodos considerados —GridSearch, RandomSearch y Optuna—, Optuna es el más eficaz. A diferencia de GridSearch, que realiza una búsqueda exhaustiva y por lo tanto es muy lento y computacionalmente costoso para grandes volúmenes de datos y numerosos hiperparámetros, y de RandomSearch, que aunque reduce el tiempo de búsqueda mediante la selección aleatoria de parámetros no garantiza la exploración eficiente del espacio de hiperparámetros; Optuna _"utiliza técnicas de optimización basadas en gradientes y búsqueda bayesiana que priorizan áreas prometedoras del espacio de parámetros"_, optimizando así tanto la calidad del modelo como el tiempo de computación necesario.


- **¿Cuál de los métodos es más rápido?**
En términos de velocidad de entrenamiento, modelos más simples como la regresión lineal o logística son más rápidos que modelos complejos como Random Forest, principalmente debido a su menor complejidad computacional y a la menor cantidad de hiperparámetros a ajustar.


- **¿Los resultados son mejores que los regresores dummy?**
Como era de esperar, los modelos de regresión más complejos superan significativamente a los regresores dummy en términos de precisión y capacidad predictiva. Los regresores dummy, que simplemente predicen la media de la variable objetivo, son únicamente útiles como punto de referencia para comparar con otros modelos más sofisticados.


- **¿El ajuste de hiperparámetros mejora con respecto a los valores por omisión?**
Como también era de esperar, el ajuste de hiperparámetros mejora significativamente el desempeño de los modelos en comparación con los valores por defecto. Los modelos optimizados, especialmente Random Forest y SVM, superan a los modelos con hiperparámetros por defecto en términos de precisión y capacidad predictiva.
Cabe destacar la increíble mejora de SVM, cuyo diferencial de RMSE y R2 entre los hiperparámetros por defecto y los ajustados es el más alto de todos los modelos.


- **¿Hay algún equilibrio entre tiempo de ejecución y mejora de resultados?**
Existe un compromiso claro entre el tiempo de entrenamiento y la precisión del modelo. Métodos de búsqueda de hiperparámetros más rápidos como RandomSearch pueden no explorar el espacio de parámetros tan exhaustivamente como Optuna, resultando en un posible deterioro de la calidad del modelo. Sin embargo, Optuna representa un buen equilibrio, ofreciendo mejoras sustanciales en el desempeño del modelo con un costo temporal relativamente moderado.


- **¿Es posible extraer de alguna técnica qué atributos son más relevantes?**
Sí, modelos como Random Forest son particularmente útiles para evaluar la importancia de los atributos. Estos modelos proporcionan _insights_ sobre cuáles variables tienen mayor impacto en la predicción a través de la medición de la importancia de las características, que calcula cómo cada atributo contribuye a la reducciónde la impureza en cada división que involucra dicho atributo.
